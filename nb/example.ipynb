```python
el_sc = pd.read_csv(
    "data/gene_score.tsv",
    sep="\t",
    index_col=0,
    squeeze=True,
)
```

```python
se_el1_ = kwat.gmt.read(["data/c2.all.v7.1.symbols.gmt"])
```
