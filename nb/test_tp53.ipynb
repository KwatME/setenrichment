```python
import numpy as np
import pandas as pd

import gsea
import kwat
```

```python
with open("data/test_p53/P53_Biocarta.Gsea.1626989856425/edb/P53.cls") as io:

    io.readline()

    io.readline()

    ta = np.array([1 - int(la) for la in io.readline().strip().split()])

# pd.value_counts(ta)
```

```python
sc_el_sa = pd.read_csv(
    "data/test_p53/P53_hgu95av2_collapsed_to_symbols.gct",
    skiprows=2,
    sep="\t",
    index_col=0,
).drop("DESCRIPTION", axis=1)

# sc_el_sa
```

```python
se_el_ = kwat.gmt.read(
    ["data/test_p53/P53_Biocarta.Gsea.1626989856425/edb/gene_sets.gmt"]
)

# len(se_el_)
```

```python
ol_el_sc = pd.read_csv(
    "data/test_p53/P53_Biocarta.Gsea.1626989856425/edb/P53_hgu95av2_collapsed_to_symbols.rnk",
    sep="\t",
    index_col=0,
    header=None,
    squeeze=True,
)

# ol_el_sc
```

```python
el_sc = gsea.compare_with_target(ta, sc_el_sa, "signal_to_noise", separate=True)

# el_sc
```

```python
co = pd.concat([ol_el_sc, el_sc], axis=1)

co.columns = ["Old rank", "New rank"]

co
```

```python
ol_en_se_st = pd.read_csv(
    "data/test_p53/P53_Biocarta.Gsea.1626989856425/gsea_report_for_WT_1626989856425.tsv",
    sep="\t",
    index_col=0,
)

# ol_en_se_st
```

```python
pr_en_se_st = gsea.run_prerank_gsea(
    ol_el_sc,
    se_el_,
    n_pe=10,
)

# pr_en_se_st
```

```python
gs_en_se_st = gsea.run_gsea(
    ta,
    sc_el_sa,
    se_el_,
    n_pe=10,
)

# gs_en_se_st
```

```python
co = pd.concat(
    [ol_en_se_st["ES"], pr_en_se_st["Enrichment"], gs_en_se_st["Enrichment"]], axis=1
)

co.columns = [
    "Old GSEA",
    "New prerank GSEA with old rank",
    "New GSEA (with new rank)",
]

co
```
